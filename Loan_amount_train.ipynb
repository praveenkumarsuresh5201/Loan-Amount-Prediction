```python
import pandas as pd

# Load your dataset into a DataFrame (df)

# List of specific columns you want to fill with the mean
columns_to_fill = ['Credit_History','Loan_Amount_Term', 'LoanAmount', 'Self_Employed', 'Dependents','Gender']

# Fill null values in specific columns with the mean of each column
df[columns_to_fill] = df[columns_to_fill].fillna(df[columns_to_fill].mean(numeric_only=True))
```

```python
# Forward fill null values in specific columns
columns_to_fill = ['Self_Employed', 'Dependents', 'Married', 'Gender']
df[columns_to_fill] = df[columns_to_fill].fillna(method='ffill')

output_file = 'train_dataset.csv'
df.to_csv(output_file, index=False)
```
